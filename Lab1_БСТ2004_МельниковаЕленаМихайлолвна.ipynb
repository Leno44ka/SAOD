### Выбором

```python
def selectionSort(matrix):
    for row in matrix:
        for i in range(len(row)):
            smallest = i
            for j in range(i + 1, len(row)):
                if row[j] < row[smallest]:
                    smallest = j
            row[i], row[smallest] = row[smallest], row[i]
            
```

### Вставкой

```python
def insertionSort(matrix):
    for g in range (len(matrix)):
        for i in range (1,len(matrix[0])):
            key=matrix[g][i]
            j=i-1
            while j>=0 and key<matrix[g][j]:
                matrix[g][j+1]=matrix[g][j]
                j -= 1
                matrix[g][j+1] = key
    return matrix
```

### Обменом

```python
def bubbleSort (matrix):
    for g in range (len(matrix)):
        for i in range (len(matrix[0])-1):
            for j in range (len(matrix[0])-i-1):
                if matrix[g][j] > matrix[g][j+1]:
                     matrix[g][j], matrix[g][j+1] = matrix[g][j+1], matrix[g][j]
    return matrix
```

### Шелла

```python
def shellSort(matrix):
    for i in range(len(matrix)):
        d=int(len(matrix[i])/2)
        while d>0:
            for j in range(len(matrix[i])):
                for h in range(int(j+d),len(matrix[i]),d):
                    if matrix[i][j]>matrix[i][h]:
                        temp=matrix[i][j]
                        matrix[i][j]=matrix[i][h]
                        matrix[i][h]=temp
            
            d=int(d/2)
    return matrix
```

### Турнирная

```python
def tournamentSort(matrix):
    for i in range(len(matrix)):
        ttournamentSort(matrix[i])
    return matrix
def ttournamentSort(matrix):
    tree = [None] * 2 * (len(matrix) + len(matrix) % 2)
    index = len(tree) - len(matrix) - len(matrix) % 2
    
    for i, v in enumerate(matrix):
        tree[index + i] = (i, v)
    
    for j in range(len(matrix)):
        n = len(matrix)
        index = len(tree) - len(matrix) - len(matrix) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) 
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        matrix[j] = x
        tree[len(tree) - len(matrix) - len(matrix) % 2 + index] = None
```

### Быстрая

```python
def quickSort(matrix):
    for i in range(len(matrix)):
        qquickSort(0,len(matrix[i])-1,matrix,i)
    return matrix

def qquickSort(_first, _last, matrix, row):
    first=int(_first)
    last=int(_last)
    middle=int((first+last)/2)
    
    while (first<last):
        
        while( matrix[row][first] < matrix[row][middle]): 
            first+=1
        while( matrix[row][last] > matrix[row][middle]):
            last-=1
        if(first<=last):
            matrix[row][first],matrix[row][last]=matrix[row][last],matrix[row][first]
            first+=1
            last-=1
    
    if(_first<last):
        qquickSort(_first,last,matrix,row)
    if(first<_last):
        qquickSort(first,_last,matrix,row)
```

### Пирамидальная

```python
def heapify(matrix, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and matrix[i] <matrix[l]:
        largest = l

    if r < n and matrix[largest] < matrix[r]:
        largest = r

    if largest != i:
        matrix[i],matrix[largest] = matrix[largest],matrix[i]

        heapify(matrix, n, largest)
def heapSort(matrix):

    for i in range(len(matrix)):
        n = len(matrix[i])

        for j in range(n, -1, -1):
            heapify(matrix[i], n, j)

        for j in range(n-1, 0, -1):
            matrix[i][j], matrix[i][0] = matrix[i][0], matrix[i][j] 
            heapify(matrix[i], j, 0)
    return matrix
```
